```python
# convert primary diagnosis code strs to floats in order to bin using numeric ranges
def bin_diag_codes(diag):
    df[diag] = np.where(df[diag].str.startswith('V'), '10'+df[diag].str[1:], df[diag])
    df[diag] = np.where(df[diag].str.startswith('E'), '1'+df[diag].str[1:], df[diag])
    df[diag] = np.where(df[diag] == '?', '2000', df[diag])
    df[diag] = df[diag].astype(np.float64)

# bin primary-df[diag]nosis-codes using numeric ranges
    conditions = [(df[diag] >= 1) & (df[diag] <= 139), (df[diag] >= 140) & (df[diag] <= 239),
                  (df[diag] >= 240) & (df[diag] <= 279), (df[diag] >= 280) & (df[diag] <= 289),
                  (df[diag] >= 290) & (df[diag] <= 319), (df[diag] >= 320) & (df[diag] <= 389),
                  (df[diag] >= 390) & (df[diag] <= 459), (df[diag] >= 460) & (df[diag] <= 519),
                  (df[diag] >= 520) & (df[diag] <= 579), (df[diag] >= 580) & (df[diag] <= 629),
                  (df[diag] >= 630) & (df[diag] <= 679), (df[diag] >= 680) & (df[diag] <= 709),
                  (df[diag] >= 710) & (df[diag] <= 739), (df[diag] >= 740) & (df[diag] <= 759),
                  (df[diag] >= 760) & (df[diag] <= 779), (df[diag] >= 780) & (df[diag] <= 799),
                  (df[diag] >= 800) & (df[diag] <= 999), (df[diag] >= 1001) & (df[diag] <= 1091),
                  (df[diag] >= 1800) & (df[diag] <= 1999), df[diag] == 2000]
    
    choices = ['Inf/Par', 'Neoplasms', 'End/Nut/Met/Imm', 'Blood', 'Mental', 'Nervous',
           'Circulatory', 'Respiratory', 'Digestive', 'Genitourinary', 'Pre/Chi/Pue',
           'Skin', 'Musculoskeletal', 'Congenital', 'Newborn', 'Ill-Defined',
           'Inj/Poi', 'Other Factors', 'Inj/Poi Supp.', 'Unknown']
    
    return np.select(conditions, choices)

df['diag_1'] = bin_diag_codes('diag_1')
df['diag_2'] = bin_diag_codes('diag_2')
df['diag_3'] = bin_diag_codes('diag_3')
```

```python
# bin discharge disposition codes using domain knowledge   
conditions = [df.discharge_disposition_id==1,
              df.discharge_disposition_id.isin([11, 19, 20, 21]),
              df.discharge_disposition_id.isin([18, 25, 26]),
              df.discharge_disposition_id.isin([2, 3, 4, 5, 9, 10, 15, 22, 23, 24]),
              df.discharge_disposition_id.isin([16, 17]),
              df.discharge_disposition_id.isin([27, 28, 29, 30]),
              df.discharge_disposition_id.isin([6, 8]),
              df.discharge_disposition_id == 7,
              df.discharge_disposition_id.isin([13, 14]),
              df.discharge_disposition_id == 12]

choices = ['Home, no Care', 'Expired', 'Unknown', 'Inpatient',
          'Outpatient', 'Unknown Status Facility', 'Home Care', 'AMA',
          'Hospice', 'To Return']

df.discharge_disposition_id = np.select(conditions, choices)
```

```python
# load grid parameters
grid_tree = pd.read_pickle('Pickles/dt_grid.pickle')

# train test split
X_train, X_test, y_train, y_test = train_test_split(le_X, y, test_size=.2, random_state=42)

# build baseline decision tree using grid search to determine optimal parameters 
dtc = DecisionTreeClassifier(max_depth=grid_tree.best_params_['max_depth'],
                             max_features=grid_tree.best_params_['max_features'],
                             min_samples_leaf=grid_tree.best_params_['min_samples_leaf'],
                             min_samples_split=grid_tree.best_params_['min_samples_split'],
                             random_state=42, class_weight = 'balanced')
# fit to the training data
model_log = dtc.fit(X_train, y_train)

# use fitted model to predict on the test data
y_hat_train = dtc.predict(X_train)
y_hat_test  = dtc.predict(X_test)

# print confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_hat_test)
print('Confusion Matrix:\n', cnf_matrix)

# visualize confusion matrix of the test set
metrics.plot_confusion_matrix(dtc, X_test, y_test,
                              display_labels=['Yes', 'No'],
                              cmap=plt.cm.Blues)
plt.grid(False)
plt.show()

print('Training Precision: ', metrics.precision_score(y_train, y_hat_train))
print('Testing Precision: ', metrics.precision_score(y_test, y_hat_test))

print('Training Recall: ', metrics.recall_score(y_train, y_hat_train))
print('Testing Recall: ', metrics.recall_score(y_test, y_hat_test))

print('Training Accuracy: ', metrics.accuracy_score(y_train, y_hat_train))
print('Testing Accuracy: ', metrics.accuracy_score(y_test, y_hat_test))

print('Training F1-Score: ', metrics.f1_score(y_train, y_hat_train))

print('Testing F1-Score: ', metrics.f1_score(y_test, y_hat_test))
```
